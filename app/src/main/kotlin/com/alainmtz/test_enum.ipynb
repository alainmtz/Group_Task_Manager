```python
enum class TaskPriority {
    LOW,
    MEDIUM,
    HIGH
}

fun parsePriority(value: String): TaskPriority {
    return try {
        TaskPriority.valueOf(value.uppercase())
    } catch (e: IllegalArgumentException) {
        // Try to handle "TaskPriority.MEDIUM" format
        if (value.contains(".")) {
            val parts = value.split(".")
            if (parts.size > 1) {
                try {
                    return TaskPriority.valueOf(parts.last().uppercase())
                } catch (e2: IllegalArgumentException) {
                    return TaskPriority.MEDIUM
                }
            }
        }
        TaskPriority.MEDIUM
    }
}

println(parsePriority("MEDIUM"))
println(parsePriority("medium"))
println(parsePriority("TaskPriority.MEDIUM"))
println(parsePriority("INVALID"))
```
